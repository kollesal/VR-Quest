## Create nex Variables that will store the mirrored values to the Y-Line

```python
# Calculate the difference between the original y-values and y-line
y_diff = yLine - df_MediaPInts.iloc[:, 1::3].values

# Mirror the values across the y-line
mirrored_values = yLine + y_diff

# Store the mirrored values in a new variable
mirrored_df = df_MediaPInts.copy()  # Create a copy of the original DataFrame
mirrored_df.iloc[:, 1::3] = mirrored_values  # Replace the y-values with mirrored values

# Display the mirrored DataFrame
mirrored_df.head(5)
```

```python
# Calculate the difference between the original y-values and y-line
y_diff = yLine - df_MediaPInts.iloc[:, 1::3]

# Mirror the y-values across the y-line
mirrored_y = yLine + y_diff

# Calculate the difference between the original x-values and x-line
x_diff = xLine - df_MediaPInts.iloc[:, ::3]

# Mirror the x-values across the x-line
mirrored_x = xLine + x_diff

# Create a DataFrame to store the mirrored points
mirrored_df = df_MediaPInts.copy()

# Replace original x-values with mirrored x-values
mirrored_df.iloc[:, ::3] = mirrored_x

# Replace original y-values with mirrored y-values
mirrored_df.iloc[:, 1::3] = mirrored_y

# Show the mirrored DataFrame
mirrored_df.head(5)
```

```python
mirrored_df['path'] = df_MediaP['path']
mirrored_df['frame'] = df_MediaP['frame']
mirrored_df['compensation'] = df_MediaP['compensation']

path_column = mirrored_df.pop('path')
frame_column = mirrored_df.pop('frame')
compensation_column = mirrored_df.pop('compensation')

mirrored_df.insert(0, 'path', path_column)
mirrored_df.insert(0, 'frame', frame_column)
mirrored_df.insert(0, 'compensation', compensation_column)

mirrored_df.head(5)
```
