```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data to a pandas data frame
df_MediaP = pd.read_csv('../Data/samples_rotated.csv', sep=',', encoding='utf-8')
df_MediaP.head(5)
```

```python
# Initialize instances to store the first values
first_value_instance_X = None
first_value_instance_Y = None

# Iterate over column names
for column in df_MediaP.columns:
    if 'x_0' in column:
        # Find the index of the first non-null value in the column
        first_value_index = df_MediaP[column].first_valid_index()
        if first_value_index is not None:
            # Extract the first value from the column
            first_value_instance_X = df_MediaP.at[first_value_index, column]
            # Break the loop once the first value is found
            break

# Iterate over column names
for column in df_MediaP.columns:
    if 'y_0' in column:
        # Find the index of the first non-null value in the column
        first_value_index = df_MediaP[column].first_valid_index()
        if first_value_index is not None:
            # Extract the first value from the column
            first_value_instance_Y = df_MediaP.at[first_value_index, column]
            # Break the loop once the first value is found
            break

# Print the first value instances
print("First value instance X:", first_value_instance_X)
print("First value instance Y:", first_value_instance_Y)
```

```python
# Find the column names containing "x"
x_columns = [col for col in df_MediaP.columns if 'x' in col]

# Iterate over the x-columns and deduct first_value_instance_X
for column in x_columns:
    df_MediaP[column] -= first_value_instance_X

# Find the column names containing "y"
y_columns = [col for col in df_MediaP.columns if 'y' in col]

# Iterate over the y-columns and deduct first_value_instance_Y
for column in y_columns:
    df_MediaP[column] -= first_value_instance_Y
    
df_MediaP.head(5)
```

```python
df_MediaP.to_csv('../Data/samples_normalised.csv',
                           sep=",",
                           encoding='utf-8',
                           index=False)
```
